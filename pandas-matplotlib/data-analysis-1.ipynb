```python
navy_columns = [
    'Aircraft Carriers', 'Corvettes', 'Destroyers', 'Frigates',
    'Helicopter Carriers', 'Mine Warfare', 'Navy Ships',
    'Patrol Vessels', 'Submarines'
]

df['Total Navy Strength'] = df[navy_columns].sum(axis=1)

df[['Total Navy Strength', 'country']]
```

```python
# save data to a new csv file
df.to_csv('modified.csv')
```

```python
df.loc[df['country'].str.contains('United')]
```

```python
# categorize countries by navy strength
def categorize(n):
    if n < 50:
        return 'Low'
    elif 50 <= n <= 200:
        return 'Medium'
    else:
        return 'High'

df['Navy Strength Category'] = df['Total Navy Strength'].apply(categorize)

navy_strength_grouped = df.groupby('Navy Strength Category').mean()

navy_strength_grouped
```

```python
import matplotlib.pyplot as plt
```

```python
top_10_defense = df.nlargest(10, 'Defense Budget')

plt.figure(figsize=(12, 6))
plt.bar(top_10_defense['country'], top_10_defense['Defense Budget'], color='skyblue')
plt.title('Top 10 countries by defense budget')
plt.xlabel('Country')
plt.ylabel('Defense Budget (in 100 Billion USD)')
plt.grid(True)
plt.show()
```

```python
plt.scatter(top_10_defense['Active Personnel'], top_10_defense['Tanks'])
for i in range(len(top_10_defense)):
    plt.text(top_10_defense.iloc[i]['Active Personnel'], top_10_defense.iloc[i]['Tanks'], top_10_defense.iloc[i]['country'], fontsize=8, ha='right')

plt.title('Active Personnel vs Tanks')
plt.xlabel('Active Personnel')
plt.ylabel('Tanks')
plt.grid(True)
plt.show()
```

```python
top_10_aircraft = df.nlargest(10, 'Total Aircraft Strength')

plt.figure(figsize=(12, 8))
plt.pie(top_10_aircraft['Total Aircraft Strength'], labels=top_10_aircraft['country'])
plt.title('Top 10 Countries by Total Aircraft Strength')
plt.show()
```
